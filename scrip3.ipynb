```python
def crawl(noun):
    print(f"noun: {noun}")
    if noun=="": 
        print("null")
        return
    response=requests.get(f"https://www.qmez.de:8444/v1/scanner/es/s?w={noun}")
    time.sleep(3)
    if response.text=='':
        print("null")
        return
    js_gender=response.json()
    # print()
    litecursor=lite.cursor()
    genders_ls.append(js_gender)
    gender=js_gender['word']
    cursor=cnx.cursor()
    litecursor.execute(f"insert into request(url,gender,noun) values ('https://www.qmez.de:8444/v1/scanner/es/s?w={gender}','{gender}','{noun}')")
    cnx.commit()
    print(f"success ")
```

```python
thread=[]
for n in range(1,len(noun)):
    print(n)
    try:
        t=threading.Thread(target=crawl,args=(noun[n],))
        thread.append(t)
        t.start()
    except: 
        print("failed")
        break
for t in thread:
    t.join()
```
